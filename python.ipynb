```python
# gussing game


# generate a random integer 1 and 100

import random
jackpot=random.randint(1,100)

guess = int(input('guess the number'))
counter = 1
while guess != jackpot:
    if guess < jackpot:
        print('worng! guess the higher')
    else:
        print('wrong! guess the lower')
    guess = int(input('guess the number'))
    counter += 1
else:
    print('correct guess')
    print('attempted',counter)
    

```

# Nested Loop

```python
for i in range(1,5):
    for j in range(1,5):
        print(i,j)
```

```python
rows = int(input('enter number of row'))

for i in range(1, rows +1):
    for j in range(1,i +1):
        print('*',end='')
    print()
```

```python
rows=int(input('enter number of  the rows'))
for i in range(1,rows+1 ):
    for j in range(1, i+1):
        print(j, end = '')
    for k in range(i-1,0,-1):
        print(k,end = '')
    print()
```

# Loop control statement

```python
for i in range(1,10):
    if i == 5:
        break
    print(i)
```

```python
lower = int(input('enter the lower range'))
upper = int(input('enter the upper range'))

for i in range(lower , upper +1):
    for j in range(2,i):
    
        if i%j == 0:
            print('whole number number',j)
            break
          
            
    else:
        print('prime number',i)
        
```

```python
# continue

for i in range(1,10):
    if i == 5:
        continue
    print(i)
    
```
