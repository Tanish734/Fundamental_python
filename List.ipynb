```python
# list functions
L = [1,2,3,4,5,0]

# len/min/max/sorted
print(len(L))
print(min(L))
print(max(L))
print(sorted(L, reverse = True))
```

```python
# count
L = [1,2,3,4,4,1,5]
L.count(1)
```

```python
# index
L = [1,2,3,4,4,1,5]
L.index(5)
```

```python
# reverse pramanet change
L = [2,1,5,7,0]
L.reverse()
print(L)
```

```python
# sort (vs sorted)
L = [2,1,5,7,0]
print(L)
print(sorted(L))
print(L)
L.sort()
print(L)
# sort is pramanet change
```

```python
# copy > shallow copy
L = [1,2,3,4,4,1,5]
print(L)
print(id(L))
L1 = L.copy()
print(L1)
print(id(L1))

```

# ways to travers a list

```python
L = [1,2,3,4]

for i in L:
    print(i)
```

```python
# index
L = [2,2,3,4]
for i in range(0,len(L)):
    print(i)
```

```python
# index
L = [2,2,3,4]
for i in range(0,len(L)):
    print(L[i])
```

```python
# Zip

L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

list(zip(L1,L2))

[i +j for i ,j in zip(L1,L2)]
```
