```python
def is_even(num):
    """
    This functions returns if a given is odd or even
    input - any valid integer
    output - odd/even
    create on - 12th nov 2025
    """
    if num == int:
        if num %2 == 0:
            return 'even'
        else:
            return 'odd'
    else:
        print('not vaild number')
```

```python
# **kwargs
# **kwargs allows us to any numbers of keyword arguments
## keyword arguments mean that they contain a key - value pair , like a pthon dictionary.

def display(**kwargs):
    for (key ,value) in kwargs.items():
        print(key ,'->', value)
```

# A lambda function is a small anonymous function.

# A lambda function can take any number of arguments, but can only have one expression.


```python
a=lambda x : x**2
a(2)
```

```python
b= lambda x,y: x+y
```

```python
b(3,4)
```

```python
#### Diff between lambda vs Normal Function

# - No name
# - lambda has no return value(infact,returns a function)
# - lambda is written in 1 line
# - not reusable

# Then why use lambda functions?<br>
# **They are used with HOF**
```

```python
# check if a string ha 'a'
a = lambda s : 'a' in s
a('hello')
```

```python
a = lambda x : 'even' if x%2 == 0 else 'odd'
```

```python
a(2)
```

```python
# higher order functions
def square(x):
    return x**2
# hof
def transform(f, L):
    output = []
    for i in L:
        output.append(f(i))
    print(output)

L = [1,2,3,4,5]
transform(square, L)
```

```python
transform(lambda x: x**3,L)
```

```python
# map
list(map(lambda x: x**2,[1,2,3,4,5]))
```

```python
L= [1,2,3,4,5]
list(map(lambda x : 'even' if x%2 == 0 else 'odd',L))
```

```python
# fetch names from a list of dict

users = [
    {
        'name':'Rahul',
        'age':45,
        'gender':'male'
    },
    {
        'name':'Nitish',
        'age':33,
        'gender':'male'
    },
    {
        'name':'Ankita',
        'age':50,
        'gender':'female'
    }
]

list(map(lambda users:users['name'],users))
```
